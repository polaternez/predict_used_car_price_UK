# Setup

```python
import numpy as np
import pandas as pd
```

# Load data
We use <a href="https://www.kaggle.com/datasets/adityadesai13/used-car-dataset-ford-and-mercedes">100,000 UK Used Car Data set</a> from Kaggle
## Content
The following data sets contains information of price, transmission, mileage, fuel type, road tax, miles per gallon (mpg), and engine size.
* audi.csv - Audi used car list
* bmw.csv - BMW used car list
* ford.csv - Ford used car list
* hyundi.csv - Hyundi used car list
* merc.csv - Mercedes used car list
* skoda.csv - Skoda used car list
* toyota.csv - Toyota used car list
* vauxhall.csv - Vauxhall used car list
* vw.csv - Volkswagen used car list

```python
audi_data = pd.read_csv("./data/raw/audi.csv")
bmw_data = pd.read_csv("./data/raw/bmw.csv")
ford_data = pd.read_csv("./data/raw/ford.csv")
hyundi_data = pd.read_csv("./data/raw/hyundi.csv")
merc_data = pd.read_csv("./data/raw/merc.csv")
skoda_data = pd.read_csv("./data/raw/skoda.csv")
toyota_data = pd.read_csv("./data/raw/toyota.csv")
vauxhall_data = pd.read_csv("./data/raw/vauxhall.csv")
vw_data = pd.read_csv("./data/raw/vw.csv")
```

```python
# column name 'tax(£)' to 'tax'
hyundi_data.columns = ['model', 'year', 'price', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg', 'engineSize']
```

```python
# add brand column all datasets
data_dict =  {"Audi": audi_data, "BMW": bmw_data, "Ford": ford_data, "Hyundi": hyundi_data, "Mercedes": merc_data,
                "Skoda": skoda_data, "Toyota": toyota_data, "Vauxhall": vauxhall_data, "Volkswagen": vw_data}

for key, data in data_dict.items():
    data["brand"] = key
```

```python
# Merge datasets
car_df = pd.concat([audi_data, bmw_data, ford_data, hyundi_data, merc_data, skoda_data, toyota_data, vauxhall_data, vw_data], axis=0)
car_df
```

```python
car_df.isnull().sum()
```

```python
# Drop year greater than 2020
car_df = car_df[car_df.year<=2020]
```

```python
# save final data
car_df.to_csv("./data/uk_used_car.csv", index=False)
```
